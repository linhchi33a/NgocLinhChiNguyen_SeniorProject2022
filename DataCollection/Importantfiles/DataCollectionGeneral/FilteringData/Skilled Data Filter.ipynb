```python
def plotGraph(figsizeX,figsizeY,blueName,redName,xlimNum,ylimNum,xLableBlue,xLabelRed):
    plt.rcdefaults()
    fig, (ax,ax1) = plt.subplots(1,2,figsize = (figsizeX,figsizeY))
    index = df.index
    values = df[blueName].values
    values.sort()
    ax.barh(index,values, align='center', height=1)
    ax.set_ylabel('Games No')
    ax.set_xlim([0,xlimNum])
    ax.set_ylim([0,ylimNum])
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel(xLableBlue)

    index = df.index
    values1 = df[redName].values
    values1.sort()
    ax1.barh(index,values1, align='center', height=1,color='r')
    ax1.set_xlim([0,xlimNum])
    ax1.set_ylim([0,ylimNum])
    ax1.invert_yaxis()  # labels read top-to-bottom
    ax1.set_xlabel(xLabelRed)
    
    fig.tight_layout()
    plt.show()
```

```python
def filter_numerical(df,numerical, numerical_value, relationship):
    if relationship == 'greater':
        df_filter = df.loc[df[numerical] > numerical_value]
    elif relationship == 'less':
        df_filter = df.loc[df[numerical] <= numerical_value]     
    return df_filter
```

```python
df_filter = filter_numerical(df,'blueWardsPlaced',150, 'less')
df_filter = filter_numerical(df_filter,'redWardsPlaced',150, 'less')
df_filter = filter_numerical(df_filter,'blueWardsDestroyed',20, 'less')
df_filter = filter_numerical(df_filter,'redWardsDestroyed',20, 'less')
df_filter = filter_numerical(df_filter,'blueTowersDestroyed',6, 'less')
df_filter = filter_numerical(df_filter,'redTowersDestroyed',6, 'less')
df_filter
```
